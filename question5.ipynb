```python
import numpy as np
import matplotlib.pyplot as plt

x_data = np.array([0.01, 0.13, 0.18, 0.21, 0.3, 0.43, 0.48, 0.61,0.68, 0.79, 0.82, 0.85, 0.9, 0.96])
y_data = np.array([0.25, 0.33, 0.35, 0.36, 0.38, 0.4, 0.41, 0.46,0.48, 0.54, 0.61, 0.62, 0.7,0.72])

x_mean = np.mean(x_data)
y_mean = np.mean(y_data)

w = np.sum((x_data - x_mean) * (y_data - y_mean)) / np.sum((x_data - x_mean) ** 2)
b = y_mean - w * x_mean

x_fit = np.linspace(0, 1, 100)
y_fit = w * x_fit + b

plt.figure(figsize=(8, 5))
plt.plot(x_fit, y_fit, 'r-')
plt.plot(x_data, y_data, 'o',  mfc='none')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear Regression example of page 15')
plt.xlim(0, 1)
plt.ylim(0, 1)

plt.show()
```
